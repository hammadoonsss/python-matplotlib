```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,7,9])
plt.show()
```

```python
plt.plot([1,2,3],[2,7,9])
plt.title("Sample Data")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')
x1 = [1,4,7,8,14,17]
y1 = [3,4,8,11,14,15]
x2 = [5,7,10,14,16]
y2 = [6,7,13,17,18]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()
```

```python
from matplotlib import style

style.use('fivethirtyeight')
x1 = [2,6,7,10,12,20,23,27]
y1 = [3,4,8,11,14,19,24,28]
x2 = [5,9,10,14,16,21,22,26]
y2 = [6,7,13,17,18,25,27,29]

plt.plot(x1,y1,label="First", linewidth=2)
plt.plot(x2,y2,label="Second")
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use('bmh')
x1 = [2,6,7,10,12,20,23,27]
y1 = [3,4,8,11,14,19,24,28]
x2 = [5,9,10,14,16,21,22,26]
y2 = [6,7,13,17,18,25,27,29]

plt.scatter(x1,y1,label="First")
plt.scatter(x2,y2,label="Second")
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use('bmh')
x1 = [2,6,7,10]
x2 = [5,9,10,10]

plt.hist(x1,label="First")
plt.hist(x2,label="Second")
plt.legend()
plt.show()
```
