### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph - 1

```python
x = [12,45,78]
y = [14,68,97]

plt.plot(x,y)

plt.title("Sample Graph", fontdict={'fontsize': 20})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

### Basic Graph - 2

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# Resize the Graph
plt.figure(figsize=(5,3), dpi=200)


#Line 1

# Keyword Argument Notation (dpi specify pixels per inch, When saving probably should use 300 if possible)
# plt.plot(x,y,label="2X",color="#ba1020", linewidth=3, linestyle="--", marker=".", markersize=17, markeredgecolor="blue")

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, "b*--", label="2X")

#Line 2

# select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)


# plt.plot(x2,x2**2,'r^-', label="X^2")

# Plot part of the graph as line 
plt.plot(x2[:6],x2[:6]**2,'r^-', label="X^2")

# Plot remainder of graph as a dot
plt.plot(x2[5:],x2[5:]**2,'r--')

# Add a title (specify the font parameter with fontdict)
plt.title("Sample Graph 2", fontdict={'fontsize': 20})

# X and Y labels
plt.xlabel("X-Line")
plt.ylabel("Y-Line")

# X and Y axis Tickmarks (Scale of the graph)
plt.xticks(x)
# plt.yticks([0,2,4,6,8,10])

# Add a Legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig("sample_graph.png", dpi=300)

# Show plot
plt.show()
```
